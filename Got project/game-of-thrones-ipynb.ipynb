```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_json('/kaggle/input/game-of-thrones-data/script-bag-of-words.json')
# This file shows that the dialogues spoken by each character in particular episodes .. 
```

```python
df.tail()
```

```python
df.head()
```

```python
df.iloc[-1]['text']
```

```python
dialogue={}
for index, row in df.iterrows():
    for item in row['text']:
        if item['name'] in dialogue:
            dialogue[item['name']]=dialogue[item['name']]+item['text']
        else:
            dialogue[item['name']]=item['text']+" "
            
```

```python
len(dialogue)
```

```python
new_df=pd.DataFrame()
new_df['character']=dialogue.keys()
new_df['words']=dialogue.values()
```

```python
new_df.iloc[:,0:3].head()
```

```python
new_df['num_words']=new_df['words'].apply(lambda x:len(x.split()))
```

```python
new_df=new_df.sort_values('num_words',ascending=False)
```

```python
new_df=new_df.head(100)
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english')
```

```python
embeddings=cv.fit_transform(new_df['words']).toarray()
```

```python
embeddings.shape
```

```python
embeddings=embeddings.astype('float64')
```

```python
from sklearn.manifold import TSNE
```

```python
tsne=TSNE(n_components=2, verbose=1,random_state=123)
z=tsne.fit_transform(embeddings)
```

```python
z.shape
```

```python
new_df['x']=z.T[0]
new_df['y']=z.T[1]
# This will show 2d representation of each character in vectors 
```

```python
new_df
```

```python
import plotly.express as px
fig=px.scatter(new_df.head(50),x="x",y="y",color="character")
fig.show()
# This graph will show similarities between characters 
```

```python
import pandas as pd
import numpy as np
import pickle

with open('got_characters_tsne.pkl', 'wb') as file:
    pickle.dump(new_df, file)


with open('got_characters_tsne.pkl', 'rb') as file:
    loaded_new_df = pickle.load(file)

print(loaded_new_df.head())
print(loaded_new_df.shape)
```
